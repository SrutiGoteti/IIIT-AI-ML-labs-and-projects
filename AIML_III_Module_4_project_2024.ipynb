## $$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x ** 2 + x + 1
```

## Slope & Derivatives

Creating a function for the derivative of $f(x)$ called df(x)

```python
def df(x):
    return 2 * x + 1
```

## Gradient Descent

```python
# ------------------------------
# TASK-1: Gradient Descent to find minimum of f(x) = x^2 + x + 1
# ------------------------------

# Initialize variables
new_x = 3               # Starting point
previous_x = 0          # Previous x value
step_multiplier = 0.1   # Learning rate
precision = 0.00001     # Threshold to stop iteration

# Lists to store values for plotting/analysis
x_list = [new_x]
slope_list = [df(new_x)]

# Gradient Descent Loop
for n in range(500):
    previous_x = new_x

    # [REQUIRED] Compute gradient at current x
    gradient = df(previous_x)

    # [REQUIRED] Update x using gradient descent formula
    new_x = previous_x - step_multiplier * gradient

    # Step size to check convergence
    step_size = abs(new_x - previous_x)

    # Append values to lists for visualization
    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED]

    # Stop if change is smaller than precision
    if step_size < precision:
        print('Loop ran this many times:', n)
        break

# Print results
print('Local minimum occurs at x =', new_x)
print('Slope or df(x) value at this point =', df(new_x))
print('f(x) value or cost at this point =', f(new_x))
```

```python
# Superimpose the gradient descent calculations on plot

plt.figure(figsize=[20, 5])

# 1 Chart: Cost function
plt.subplot(1, 3, 1)

plt.xlim(-3, 3)
plt.ylim(0, 8)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.plot(x_1, f(x_1), color='blue', linewidth=3, alpha=0.8)

values = np.array(x_list)
plt.scatter(x_list, f(values), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 3, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('df(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

# 3 Chart: Derivative (Close Up)
plt.subplot(1, 3, 3)

plt.title('Gradient Descent (close up)', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.grid()
plt.xlim(-0.55, -0.2)
plt.ylim(-0.3, 0.8)

plt.plot(x_1, df(x_1), color='skyblue', linewidth=6, alpha=0.8)
plt.scatter(x_list, slope_list, color='red', s=300, alpha=0.6)


plt.show()
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

# Example 3 - Divergence and Overflow

## $$h(x) = x^5 - 2x^4 + 2$$

```python
# ------------------------------
# Example 3: Divergence and Overflow (Safe Version)
# Function: h(x) = x^5 - 2x^4 + 2
# ------------------------------

import numpy as np
import matplotlib.pyplot as plt

# Generate X values
x_3 = np.linspace(-1, 3, 1000)

# Define the function h(x)
def h(x):
    return x**5 - 2*x**4 + 2

# Define the derivative dh(x)
def dh(x):
    return 5*x**4 - 8*x**3

# ------------------------------
# Safe Gradient Descent Function
# ------------------------------

def gradient_descent_safe(derivative_func, initial_guess, multiplier=0.001, precision=0.0001, max_iter=500, max_x=1e6):
    """
    Performs gradient descent with overflow protection.
    Stops if x becomes too large (numerical safety).
    """
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        # Overflow protection
        if abs(new_x) > max_x:
            print(f"Stopped gradient descent: |x| too large at iteration {n}, x={new_x}")
            break

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break

    return new_x, x_list, slope_list

# ------------------------------
# Run Gradient Descent for Multiple Initial Guesses
# ------------------------------

initial_guesses = [-0.5, 0.5, 1.0, 1.5, 2.0]

for guess in initial_guesses:
    local_min, x_path, slope_path = gradient_descent_safe(
        dh, initial_guess=guess, multiplier=0.0005, precision=0.0001
    )
    print(f"Initial guess {guess} -> Local min at: {local_min}, Steps: {len(x_path)}")

    # Plot function and derivative with gradient descent path
    plt.figure(figsize=[12, 5])

    # Function
    plt.subplot(1, 2, 1)
    plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)
    plt.scatter(x_path, h(np.array(x_path)), color='red', s=50, alpha=0.6)
    plt.title(f"h(x) with Gradient Descent Path (Initial guess = {guess})", fontsize=15)
    plt.xlabel('X')
    plt.ylabel('h(x)')
    plt.grid(True)

    # Derivative
    plt.subplot(1, 2, 2)
    plt.plot(x_3, dh(x_3), color='skyblue', linewidth=3, alpha=0.8)
    plt.scatter(x_path, slope_path, color='red', s=50, alpha=0.6)
    plt.title(f"Derivative dh(x) with Gradient Descent Path (Initial guess = {guess})", fontsize=15)
    plt.xlabel('X')
    plt.ylabel("dh(x)")
    plt.grid(True)

    plt.show()
```

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
# Function f(x, y)
def f_2d(X, Y):
    return 13 - X**2 - Y**2 + 1

# Gradient of f(x, y)
def grad_f(X, Y):
    df_dx = -2 * X
    df_dy = -2 * Y
    return np.array([df_dx, df_dy])
```

```python
def gradient_descent_2d(grad_func, initial_point, learning_rate=0.1, precision=1e-6, max_iter=500):
    point = np.array(initial_point, dtype=float)
    points_list = [point.copy()]

    for i in range(max_iter):
        grad = grad_func(point[0], point[1])
        new_point = point - learning_rate * grad
        points_list.append(new_point.copy())

        # Stop if movement is smaller than precision
        if np.linalg.norm(new_point - point) < precision:
            break
        point = new_point

    return point, np.array(points_list)
```

```python
# Initial guess
initial_point = [2.5, 2.5]

# Run gradient descent
min_point, path = gradient_descent_2d(grad_f, initial_point, learning_rate=0.1)

print("Local minimum at:", min_point)
print("Number of steps:", len(path))
```

```python
# Create meshgrid
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = f_2d(X, Y)

# 3D Surface Plot
fig = plt.figure(figsize=[12, 6])
ax = fig.add_subplot(121, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax.scatter(path[:, 0], path[:, 1], f_2d(path[:, 0], path[:, 1]), color='red', s=50)
ax.set_title('Gradient Descent Path on 3D Surface')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Contour Plot
ax2 = fig.add_subplot(122)
cont = ax2.contourf(X, Y, Z, cmap='viridis', levels=50)
ax2.plot(path[:, 0], path[:, 1], color='red', marker='o', markersize=3, label='Path')
ax2.set_title('Gradient Descent Path on Contour')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.legend()
fig.colorbar(cont, ax=ax2)
plt.show()
```

## Batch Gradient Descent with SymPy

```python
from sympy import symbols, diff
import numpy as np

# Step 1: Define symbolic variables
a, b = symbols('x y')

# Step 2: Define the 2D symbolic function
f_xy_sym = 13 - a**2 - b**2 + 1  # Example function

# Step 3: Compute partial derivatives
df_dx = diff(f_xy_sym, a)
df_dy = diff(f_xy_sym, b)

# Step 4: Gradient Descent setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0], dtype=float)  # initial guess
gradients_list = []

for n in range(max_iter):
    # Evaluate derivatives at current point
    gradient_x = float(df_dx.evalf(subs={a: params[0], b: params[1]}))
    gradient_y = float(df_dy.evalf(subs={a: params[0], b: params[1]}))

    gradients = np.array([gradient_x, gradient_y])
    gradients_list.append(gradients)

    # Update parameters
    params = params - multiplier * gradients

# Step 5: Results
print('Final gradient:', gradients)
print('Minimum occurs at x =', params[0])
print('Minimum occurs at y =', params[1])
print('Cost at minimum:', float(f_xy_sym.evalf(subs={a: params[0], b: params[1]})))
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.

def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
